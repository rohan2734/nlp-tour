```python
nlp=spacy.load('en_core_web_lg',disable=['parser','tagger','ner'])
```

```python
def seperate_punc(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n']
    # ''\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ' is for ignoring \n new lines, periods (.),etc
```

```python
def create_model(vocabulary_size, seq_len):
    model = Sequential()
    model.add(Embedding(vocabulary_size, seq_len, input_length=seq_len))
    model.add(LSTM(50, return_sequences=True,input_shape=(1, 1)))
    model.add(LSTM(50))
    model.add(Dense(50, activation='relu'))

    model.add(Dense(vocabulary_size, activation='softmax'))
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
   
    model.summary()
    
    return model
```

```python
model = create_model(vocabulary_size+1,seq_len)
#vocabulary_Size is nuber of unqie words
```

```python
#train and fit model
from pickle import dump,load
```

```python
model.fit(X,y,batch_size=128,epochs=2,verbose=1)
#batch_Size, is number of sequences,128 (arbitarily)
#2 epochs is not enough, actually
#verbose is output report
```

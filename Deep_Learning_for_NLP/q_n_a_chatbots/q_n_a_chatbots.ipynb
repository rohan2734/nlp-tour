```python
#create a function that helps in vectorising the stories
```

```python
#we need max story len and max question len
#since we need to adjust the data by adding 0s or cutting it down for the padding sequences
def vectorize_stories(data,word_index=tokenizer.word_index,max_story_len=max_story_len,max_question_len=max_question_len):
    #stories = X
    X =[]
    # questions = Xq
    Xq = []
    #target(correct answer) (yes/no)
    Y=[]
    
    for story,query,answer in data:
        #for each story 
        # [23,14,....]
        x=[word_index[word.lower()] for word in story]
        xq=[word_index[word.lower()] for word in query]
        
        y = np.zeros(len(word_index) + 1)
        #index 0 is reserved for pad sequences 
        
        y[word_index[answer]]=1
        
        X.append(x)
        Xq.append(xq)
        Y.append(y)
        
    return (pad_sequences(X,maxlen=max_story_len),pad_sequences(Xq,maxlen=max_question_len),np.array(Y))
```

```python
#next we will create the model with keras layers
from keras.models import Sequential,Model
```

```python
from keras.layers.embeddings import Embedding
```

```python
from keras.layers import Input,Activation,Dense,Permute,Dropout,add,dot,concatenate,LSTM
```

```python
#we need to hold up in placeholder
#we also have stores and quesitons
#we have stories , which encoder needs to understaand 
#then a seperate queston, and we need to link them together to provide a label, 'yes' or 'no'
```

```python
#we are going to create placeholders using  input
#the input funciton of keras, is going to instantiate a tensor
input_sequence = Input((max_story_len,))
#and we have to add the shape , so we add max_story_len and we are saying ,
#we added ","  since PLACEHOLDER shape = (max_story_len,batch_size)
#since we dont know the size of the batch_Size, we have left it (or we didnt define it)
question = Input((max_question_len,))
#these are placeholdrers ready to recive input of  stories and questions
```

```python
#now we need to create input encoders
#we need to define vocabulary size
vocab_size = len(vocab) + 1
#vocab_len
```

```python
#we  need to continue for  INOUT ENCODER M
input_encoder_m = Sequential()
#this input gets embedded into sequence of vectors
input_encoder_m.add(Embedding(input_dim=vocab_size,output_dim=64))
#we are adding embedding layer, we are giving vocab size, and output_dim
input_encoder_m.add(Dropout(0.3))
#add a dropout layer,percentage of nuerons are dropped of

#OUTPUT
#{sample,story_maxlen,embedding_dim}
```

```python
#we need to copy and paste for input encoder C 

input_encoder_c = Sequential()
#this input gets embedded into sequence of vectors
input_encoder_c.add(Embedding(input_dim=vocab_size,output_dim=max_question_len))
#we are adding embedding layer, we are giving vocab size, and output_dim
input_encoder_c.add(Dropout(0.3))
#add a dropout layer,percentage of nuerons are dropped of

#OUTPUT
#{sample,story_maxlen,embedding_dim}
```

```python
#question encoder

question_encoder = Sequential()
#this input gets embedded into sequence of vectors
question_encoder.add(Embedding(input_dim=vocab_size,output_dim=64,input_length=max_question_len))
#we are adding embedding layer, we are giving vocab size, and output_dim
question_encoder.add(Dropout(0.3))
#add a dropout layer,percentage of nuerons are dropped of

#OUTPUT
#{samples,query_maxlen,embedding_dim}
```

```python
#now that we have the input encoder m, and inputencoder c and quesiton encoder
```

```python
#we need to encode the sequences 
```

```python
#we alreadyhave placeholders (input_sequence, question)
#we need to pass the placeholders into the input encoders m, c and question encoder
```

```python

#result after passing through the encoder

#encoded <---- encoder(input)
input_encoded_m = input_encoder_m(input_sequence)
input_encoded_c=input_encoder_c(input_sequence)
question_encoded=question_encoder(question)
```

```python
#we just need to use the dot product to compute the match b/w the first input sequence vector 
#and the query
match = dot([input_encoded_m,question_encoded],axes=(2,2))
#as a list, we pass the two eleemnts which we want to compare
match=Activation('softmax')(match)
```

```python
response = add([match,input_encoded_c])
response=Permute((2,1))(response)
```

```python
#once we have a response, we can concatenate the match matrix with question matrix sequence
```

```python
answer = concatenate([response,question_encoded])
```

```python
answer
```

```python
#we want to reduce it with a recurrent layer, we  are using LSTM layer specifically for it
```

```python
from keras.layers.recurrent import LSTM
answer = LSTM(32)(answer)
```

```python
#we are performing one more series of regulrisation with dropout
answer = Dropout(0.5)(answer)
answer = Dense(vocab_size)(answer)
#(sample,vocab_size) #YES/NO 0000
```

```python
answer = Activation('softmax')(answer)
```

```python
# we will need to output a proability distribution over the vocabulary, because we see , a bunch of 0 , except some probabilites on yes /no
#so we will pass them into softmax,to turn them into 0 or 1
answer = Activation('softmax')(answer)
```

```python
model = Model([input_sequence,question],answer)
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy')
```

```python
#we are doing this compilation accros entire voacbulary,
#we should expect only yes/no
```

```python
#now we need to fit the model
history = model.fit([inputs_train,queries_train],answers_train,batch_size=32,epochs=3,validation_data=([inputs_test,queries_test],answers_test))
#we are training on input stories and input questions
#100 epochs ~ 6000 seconds
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
